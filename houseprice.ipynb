```python
# Function to detect and handle outliers using the IQR method
def handle_outliers(df, numeric_columns):
    """
    Detect and handle outliers in the specified numeric columns using the IQR method.
    Outliers will be removed from the dataset.

    Parameters:
    - df: Pandas DataFrame
    - numeric_columns: List of numeric column names to check for outliers

    Returns:
    - df_cleaned: DataFrame with outliers removed
    """
    df_cleaned = df.copy()

    for column in numeric_columns:
        Q1 = df_cleaned[column].quantile(0.25)  # 25th percentile
        Q3 = df_cleaned[column].quantile(0.75)  # 75th percentile
        IQR = Q3 - Q1  # Interquartile Range

        # Define outlier bounds
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Filter out rows with outliers
        df_cleaned = df_cleaned[
            (df_cleaned[column] >= lower_bound) &
            (df_cleaned[column] <= upper_bound)
        ]



    return df_cleaned

# List of numeric columns (excluding 'Location' as it is categorical)
numeric_columns = df1.select_dtypes(include=['int64']).columns.tolist()

# Apply the outlier handling function
cleaned_data = handle_outliers(df1, numeric_columns)

# Save the cleaned dataset
df2=cleaned_data
df2.head()
df1=df2


```

```python
def prediction(location, bhk, bath, balcony, sqft):

    loc_index, area_index, avail_index = -1,-1,-1

    if location!='other':
        loc_index = int(np.where(X.columns==location)[0][0])



    x = np.zeros(len(X.columns))
    x[0] = bath
    x[1] = balcony
    x[2] = bhk
    x[3] = sqft

    if loc_index >= 0:
        x[loc_index] = 1
    if area_index >= 0:
        x[area_index] = 1
    if avail_index >= 0:
        x[avail_index] = 1

    return model.predict([x])[0]
```
